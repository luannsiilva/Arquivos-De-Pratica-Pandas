```python
# importando a biblioteca pandas
import pandas as pd

# criando dataframes dos arquivos
vendas_df = pd.read_csv('Contoso - Vendas - 2017.csv', sep=';', encoding='ISO-8859-1')
produtos_df = pd.read_csv('Contoso - Cadastro Produtos.csv', sep=';', encoding='ISO-8859-1')
lojas_df = pd.read_csv('Contoso - Lojas.csv', sep=';', encoding='ISO-8859-1')
clientes_df = pd.read_csv('Contoso - Clientes.csv', sep=';', encoding='ISO-8859-1')

# Renomeando tabelas
produtos_df = produtos_df.rename(columns={'ÿNome do Produto': 'Nome do Produto'})
lojas_df = lojas_df.rename(columns={'ÿID Loja': 'ID Loja'})
clientes_df = clientes_df.rename(columns={'ÿID Cliente': 'ID Cliente'})

# Selecionando apenas colunas nescessárias nos dataframes
produtos_df = produtos_df[['Nome do Produto', 'ID Produto']]
lojas_df = lojas_df[['Nome da Loja', 'ID Loja']]
clientes_df = clientes_df[['E-mail', 'ID Cliente']]

# Mesclando dataframes 
vendas_df = vendas_df.merge(produtos_df, on='ID Produto')
vendas_df = vendas_df.merge(lojas_df, on='ID Loja')
vendas_df = vendas_df.merge(clientes_df, on='ID Cliente')
```

```python
# Criando uma variável e armazenando a contagem de clientes dentro dela
frequencia_clientes = vendas_df['E-mail'].value_counts()

# Exibindo a variavel de contagem dentro de um gráfico
frequencia_clientes[:5].plot(figsize=(15, 5))
print(frequencia_clientes[:5])
```

```python
# Criando uma variável e dentro dela agrupando os valores e somando-os
total_quantidades_vendidas = vendas_df.groupby('Nome da Loja').sum()

#filtrando apenas a coluna de Quantidade Vendida
total_quantidades_vendidas = total_quantidades_vendidas[['Quantidade Vendida']]

# Ordenando as colunas em ordem decrescente
total_quantidades_vendidas = total_quantidades_vendidas.sort_values('Quantidade Vendida', ascending=False)

# Exibindo os valores num gráfico
total_quantidades_vendidas[:5].plot(figsize=(15, 5))
total_quantidades_vendidas[:5]
```

```python
# Pegando o maior valor de quantidade vendida
maior_quantidade_vendida = total_quantidades_vendidas['Quantidade Vendida'].max()
```

```python
# Pegando o menor valor de quantidade vendida
menor_quantidade_vendida = total_quantidades_vendidas['Quantidade Vendida'].min()
```
